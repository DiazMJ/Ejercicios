#### 2.4.3 Haz que todos los DATOS estén en minúsculas. Para ello utiliza .str.lower()
Para ello vamos a iterar por cada columna y que intente (try) cambiar los valores a minúsculas, si no puede (except) que siga con la iteración

```python
for col in Df_total.columns.to_list(): #convierto las columnas en listas para poder iterar
    try:
      Df_total[col] = Df_total[col].str.lower() #ojo no te olvides de indicar las col !!!!!!!muchacha
    except: #en el caso q no se pueda que pase al siguiente
       pass
Df_total.head()
```

#### 2.4.4 Crea nuevas columnas para el mes('contact_month') y el año('contact_year') de la columna 'date'. 
- Utiliza str.split() para indicar en qué parte del sting quieres cortar
- Luego selecciona las posiciones de la lista que nos interesa 

Para ello partiremos la columna date por los guiones, esto nos generará una lista con el día (0), mes(1) y año (2). Sólo usaremos estos dos últimos valores.

```python
# Primero creamos una variable que se llame fecha donde guardaremos el nuevo DF generado al usar str.split
Df_total['date'] #vemos como son los datos de la columna date de nuestro Df total para ver como partimos los datos, como he comentado antes vamos a partir por el guion

```

```python
df_fecha = Df_total['date'].str.split('-', expand = True)#hemos dividido la fecha de la columna date en un data frame de fecha con  el dia mes y año por separado , ya que en la columna estaban unidos por guiones. lo hemos separado con el str.split() quitando los guiones
df_fecha #vemos como ha quedado
```

```python
# Para el mes usaremos los valores en la posición 1
Df_total['contact_month'] = df_fecha[1]
# Para el año los valores en la posición 2
Df_total['contact_year'] = df_fecha[2]
#comprobamos que estan creadas
Df_total.head(6)
```

#### 2.4.5 Convierte la columna 'date' a tipo datetime.
- Primero crea un diccionario con los nombres de los meses y el número del mes.
- Ahora reemplaza los datos de 'date' usando .replace()
- Por último usa pd.todatetime para cambiar el tipo de dato de la columna (Recuerda usar el formato %d-%m-%Y)


```python
meses = {'enero':'1',
         'febrero':'2',
         'marzo':'3',
         'abril': '4',
         'mayo':'5',
         'junio':'6',
         'julio':'7',
         'agosto':'8',
         'septiembre':'9',
         'octubre':'10',
         'noviembre':'11', 
         'diciembre':'12'}


Df_total.replace({'date': meses },regex=True,inplace=True) 
```

```python
# Utilizamos pd.to_datetime para cambiar el tipo de dato
Df_total['date'] = pd.to_datetime(Df_total['date'],format = "%d-%m-%Y") 
Df_total.head(2)
```

#### 2.4.6 Reemplaza los . por espacios en la columna 'education'. Utiliza.str.replace()
En este caso, la columna 'education' está usando puntos como separadores. Vamos a cambiar eso por espacios

```python
Df_total['education']=Df_total['education'].str.replace('.', ' ')
Df_total['education']
```

#### 2.4.7 Convierte a float las columnas que deberían serlo
- Primero identifica cuáles son estas columnas y crea una lista con los nombres de estas columnas.
- Ahora itera en esa lista con un bucle for.
- Dentro del bucle utiliza str.replace() para cambiar las ',' por '.'.
- Aún dentro del bucle utiliza un apply con una función lambda en la que cambiemos el tipo de dato a float.

```python
Df_total.info()
```

```python
#he identificado las columnas cuyo tipo de datos no concuerdan y deberian ser float. Creo la lista que me pide el ejercicio.
listaFloat = {'cons_price_idx','cons_conf_idx','euribor3m'}
#ahora voy a iterar por las columnas de la lista para cambiar las comas de los datos por puntos y con lamba aplico la fx float para cambiar el tipo de str a float
for col in listaFloat:
  Df_total[col] = Df_total[col].str.replace(',','.')
  Df_total[col] = Df_total[col].apply(lambda x :float(x)) 

```

```python
Df_total.info()
```

#### 2.4.8 Mapear los valores 0 y 1 por No y Sí.
- Creamos un diccionario con claves 0 y 1 y valores 'No' y 'Sí'.
- Crea una lista con las columnas que vamos a mapear
- Itera con un bucle for por la lista y usa el método map() para cada columna
Para ello primero crearemos un diccionario con claves 0 y 1 y valores No y Sí. Después seleccionaremos las columnas a mapear y sustituiremos los valores

```python
Dic_map = {0 :'si', 1:'no' }
mapear = ['housing','loan',]
for col in mapear:
  Df_total[col] = Df_total[col].map(Dic_map)
```

```python
Df_total.head(2)
```

#### 2.4.9 Genera nulos en la columna 'pdays'. Utiliza .replace() y sustituye los valores que consideres por nulos.
No tiene mucho sentido que hayan pasado 999 días desde elo último contacto, probablemente es la forma de indicar que no tienen ese valor, es decir, es un valor nulo. Por lo tanto vamos a sustituirlo. 

```python
Df_total['pdays']=Df_total['pdays'].replace(999,np.nan) 
```
