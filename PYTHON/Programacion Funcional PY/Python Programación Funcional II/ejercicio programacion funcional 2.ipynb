````python
def procesar_datos(operacion,*args,**kwargs):
    if operacion == "sumar":
        suma = kwargs.get('valor_inicial',0)# kwargs va a buscar en un dicc la clave 'valor iicial' si la encuentra cográ su valor, si no encuntra esa clave el valor = 0
        for dato in args: # los args son tipo lista por lo tanto iterables si que buscamos el dato en la lista para operar
            suma += dato # (suma = suma + dato)
        return suma
    elif operacion == "restar":
        resta = args[0] # debemos empezara a retar por la posicion 0 de la lista ya que es nuestro valos inicnil
        for dato in args[ 1:]: # ponemos 1 : para indicar q empiece en 1 hasta el final de la lista , si ponemos 1:8 llegará hasta la posicion 8
            resta -= dato
        return resta
    elif operacion == "multiplicar":
        multiplicacion = kwargs.get('factor', 1) #utilizamos kwargs xq el enunciado nos vuelve a pedir si hay un factor multiplicacion
        for dato in args:
            multiplicacion *= dato
        return multiplicacion
    elif operacion == "dividir":
        division = args[0] # Inicializamos con el primer valor
        try:
            for dato in args[1:]:
                division /= dato
            return division
        except ZeroDivisionError:
            return "no es posible dividir entre 0"
    else:
        """raise ValueError("operacion no válida{}".format(operacion)) es lo mismo que lotro format"""
        raise ValueError(f"operacion no valida: {operacion}")

resultado_division = round(procesar_datos('dividir', 100, 5, 2))
print(f"El resultado de la division es : {resultado_division}")

"""resultado_multiplicacion = procesar_datos('multiplicar', 2, 3, 4,factor =2)
print(f"El resultado de la multiplicacion es: {resultado_multiplicacion}")"""

"""resultado_suma = procesar_datos('sumar', 1, 2, 3, 4, valor_inicial=10)
print(resultado_suma)"""
"""resultado_resta = procesar_datos('restar', 20, 5, 3)
print(f"El resultado de la resta es :{resultado_resta}")"""
````
